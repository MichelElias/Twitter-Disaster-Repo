```python
import pandas as pd
import numpy as np
import json
import pickle
import csv
```

```python
twitter_files = !ls ./data/raw_json_test2/twitter_2018*
```

```python
all_tweets = []

for file in twitter_files:
    tweets = []
    with open(file, 'r') as f:
        curr_json = json.load(f) 
    for tweet in curr_json:
        try:
            tweets.append((tweet['user']['screen_name'], 
                           tweet['created_at'],
                           tweet['user']['location'], 
                           tweet['coordinates'], 
                           tweet['geo'], 
                           tweet['text']))
        except:
            pass
    all_tweets.extend(tweets)
```

```python
df_tweets = pd.DataFrame(all_tweets, 
                         columns = ['user', 
                                    'created_at', 
                                    'user_location', 
                                    'tweet_coordinates', 
                                    'tweet_geo', 
                                    'text'])
```
